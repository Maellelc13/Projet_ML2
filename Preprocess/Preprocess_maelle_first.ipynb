```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
df = pd.read_csv("../data/kaggle_b2_fraud_train_v3.csv")
```

```python
df_clean = df.dropna().select_dtypes('number')

X = df_clean.drop('target_is_fraud', axis=1)
y = df_clean['target_is_fraud']

model = LogisticRegression(max_iter=1000)
model.fit(X, y)
 
predictions = model.predict(X)

submission = pd.DataFrame({
    'customer_id': df_clean['customer_id'], 
    'target_is_fraud': predictions
})

submission.to_csv('submission_train.csv', index=False)
print(submission.head())
```

```python
import numpy as np

# 2. Sauvegarder customer_id
customer_ids = df['customer_id']

# 3. Garder seulement les colonnes numériques pour le modèle
df_numeric = df.select_dtypes(include=[np.number])

# 4. Au lieu de dropna(), remplir les valeurs nulles avec la médiane
df_filled = df_numeric.fillna(df_numeric.median())

print("Taille avant remplissage:", df_numeric.shape)
print("Taille après remplissage:", df_filled.shape)  # 160000 lignes !

# 5. X et y
X = df_filled.drop('target_is_fraud', axis=1)
y = df_filled['target_is_fraud']

# 6. Modèle
model = LogisticRegression(max_iter=1000)
model.fit(X, y)

# 7. Prédictions
predictions = model.predict(X)

# 8. DataFrame final (160000 lignes !)
submission = pd.DataFrame({
    'customer_id': customer_ids,
    'target_is_fraud': predictions
})

print(f"Shape finale: {submission.shape}")
print(submission.head())
submission.to_csv('submission_train.csv', index=False)
```
