```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
df_train = pd.read_csv("../data/kaggle_b2_fraud_train_v3.csv")
df_test = pd.read_csv("../data/kaggle_b2_fraud_test_v3.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression

test_ids = df_test['customer_id']

df_train_numeric = df_train.select_dtypes(include=[np.number])
df_train_filled = df_train_numeric.fillna(df_train_numeric.median())

print("Train - taille après remplissage:", df_train_filled.shape)

X_train = df_train_filled.drop('target_is_fraud', axis=1)
y_train = df_train_filled['target_is_fraud']

df_test_numeric = df_test.select_dtypes(include=[np.number])
df_test_filled = df_test_numeric.fillna(df_train_numeric.median())

print("Test - taille après remplissage:", df_test_filled.shape)

X_test = df_test_filled  

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

predictions = model.predict(X_test)


submission = pd.DataFrame({
    'customer_id': test_ids,  
    'target_is_fraud': predictions
})

print(f"Shape finale de la submission: {submission.shape}")
print(submission.head())
submission.to_csv('submission_kaggle_1.csv', index=False)
```
